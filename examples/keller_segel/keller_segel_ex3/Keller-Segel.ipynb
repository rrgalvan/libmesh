### Profundizando

- Estos parámetros, que definirán el experimento numérico para el programa `example-opt`, se definen de tres formas distintas:
  1. Dentro del código C++ (en el fichero [`keller_segel_ex3.C`](./keller_segel_ex3.C)
  2. O bien dentro del fichero [`keller_segel_ex3.in`](./keller_segel_ex3.in)
  3. O bien como argumentos que se pasan al programa a través de consola
- Los parámetros por consola tienen mayor precedencia que los del fichero `.in`, que a su vez tienen mayor precedencia que los datos definidos en el código C++.

Por ejemplo, si quisiéramos ejecutar de nuevo el programa pero realizando sólo 2 iteraciones de tiempo con paso 0.01, podríamos a) modificar y recompilar el programa o b) cambiar los datos que están en el fichero [`keller_segel_ex3.in`](./keller_segel_ex3.in) o bien c) escribir lo siguiente (cuidado, parece que no se pueden dejar espacios en blanco junto al signo `=`):

```python
!./example-opt nt=2 dt=0.01
```
